```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
def DFT_equation_implement(x):

    x = np.asarray(x, dtype=float)
    N = x.shape[0]
    n = np.arange(N)
    k = n.reshape((N, 1))
    P = np.exp(-2j * np.pi * k * n / N)
    X = np.dot(P, x)
    X = [round(y.real,4) + round(y.imag,4)*1j for y in X]
    return X
```

```python
 f = 5  
fs = 50  

t = np.linspace(0, 2, 2 * fs, endpoint=False)

x = np.sin(2*np.pi*f*t) # Signal
X= DFT_equation_implement(x)
print("Pyton library and equation does the same : ",np.allclose(DFT_equation_implement(x), np.fft.fft(x)))
```

```python
plt.plot(t, x)
plt.xlabel('Time ')
plt.ylabel('Signal amplitude')
plt.title('Sample signal')
plt.grid(True)
plt.show
```

```python
fqs = np.fft.fftfreq(np.size(x)) * fs

plt.stem(fqs, np.abs(X))
plt.xlabel('Frequency in Hertz')
plt.ylabel('Frequency Domain ')
plt.title('DFT of signal x')
plt.grid(True)
plt.show
```
