```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = [10,5]
plt.rcParams.update({'font.size':8})
```

```python
ds   = 0.0001
t           = np.arange(start=0,stop=1,step=ds)
clean     = np.cos(2.5*np.pi*150*t) + np.cos(2*np.pi*120*t)
noise     = clean + 1.5*np.random.randn(len(t))
```

```python
plt.plot(t,noise,color='b',Linewidth=2,label='Noisy')
plt.plot(t,clean,color='g',Linewidth=2,label='Clean')
plt.title('Clean + noise')
plt.legend()
```

```python
from scipy.fft import rfft,rfftfreq
n    = len(t)
yf   = rfft(noise)
xf   = rfftfreq(n,ds)
plt.title('The frequency domain data')
plt.plot(xf,np.abs(yf))
```

```python
clean     = np.cos(2.5*np.pi*150*t) + np.cos(2*np.pi*120*t)
```

```python
yf_abs      = np.abs(yf) 
indices     = yf_abs>300   
yf_clean    = indices * yf 
plt.title('Clean up to the noise')
plt.plot(xf,np.abs(yf_clean))
```

```python
from scipy.fft import irfft
new_f_clean = irfft(yf_clean)
plt.plot(t,new_f_clean)
plt.title('Clean wave')
plt.ylim(-4,4)
```
